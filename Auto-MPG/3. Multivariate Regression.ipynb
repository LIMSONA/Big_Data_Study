# Auto-MPG 다중 회귀분석(Multivariate Regression)

> 여러 개의 독립변수를 적용하여 분석을 해보자!

# 라이브러리 불러오기

```python
import pandas as pd
import numpy as np

import os

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.listdir()
```

# 데이터 불러오기 & 전처리 하기

```python
# 벌써 알고있는 데이터이기에 얼른 전처리 빠박하기!
df = pd.read_csv('auto-mpg.csv',header=None)
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
```

```python
# horsepower 열의 결측치 변경 및 처리하기
df['horsepower'].replace('?', np.nan, inplace=True)      # '?'을 np.nan으로 변경
df.dropna(subset=['horsepower'], axis=0, inplace=True)   # 누락데이터 행을 삭제

#자료형 변경 (문자열 ->숫자)
df['horsepower'] = df['horsepower'].astype('float')      # 문자열을 실수형으로 변환
```

# 독립변수(X), 종속변수(y) 설정하기

## 필요없는 열 제거하기 (상관관계 확인 후!)

```python
sns.heatmap(df.corr(),annot=True, linewidths=0.5)
```

```python
# 필요한 열을 ['cylinders', 'horsepower', 'weight'] 로 정하고
# X와 y 정하기

X= df[['cylinders', 'horsepower', 'weight']]
y= df.mpg
```

## Train / Test 나누기

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y,
                                                 test_size=0.3,
                                                 random_state=10)
```

```python
print('훈련 데이터: ', X_train.shape)
print('검증 데이터: ', X_test.shape)   
```

# 모형 학습하기

```python
from sklearn.linear_model import LinearRegression

lr= LinearRegression()
```

```python
# 회귀 학습시키기
lr.fit(X_train, y_train)
```

```python
# 학습을 마친 모형에 test data를 적용하여 결정계수(R-제곱) 계산
r_square = lr.score(X_test, y_test)

print(r_square)
```

R2 결정계수 값이 약 0.7정도 나왔으니 일단 반절을 넘었으니 비교적 괜찮은 점수

```python
# 회귀식의 기울기는??
lr.coef_
```

```python
# 회귀식의 y절편(bias)은??
lr.intercept_
```

```python
# train 데이터의 scatter와 test 데이터로 회귀선을 그려보장!

y_hat = lr.predict(X_test)
```

```python
plt.figure(figsize=(9,4))

ax1 = sns.distplot(y_test, label="y_test", hist=False)
ax2 = sns.distplot(y_hat, label="y_hat", ax=ax1, hist=False)
plt.legend()
plt.show()
```

> 아래의 weight만 사용한 단순회귀 분석결과와 위의 다중회귀분석을 비교해보면 뾰족했던 편향되는 경향이 좀더 감소한 것을 알수 있다!
